```python
#import pandas numpy plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#insatll quandl
!pip install quandl
import quandl
```

```python
data = quandl.get('NSE/TATAGLOBAL')
```

```python
data.head()
```

```python
data['open-close'] = data['Open'] - data['Close']
data['high-low'] = data['High'] - data['Low']
data = data.dropna()
```

```python
X = data[['open-close', 'high-low']]
X.head()
```

```python
#classification problem wether to buy or sell(+1 for by and -1 for sell)
y = np.where(data['Close'].shift(-1) > data['Close'], 1, -1)
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#importing KNN, neighbor, GridsearchCV, accuracyscore
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

#using GridsearchCV to find best parameter
params = {'n_neighbors': np.arange(1, 20)}
knn = KNeighborsClassifier()
model = GridSearchCV(knn, params, cv=5)
model.fit(X_train, y_train)

#accuracyscore
accuracy_train = accuracy_score(y_train, model.predict(X_train))
accuracy_test = accuracy_score(y_test, model.predict(X_test))
print('Train Accuracy: ', accuracy_train)
print('Test Accuracy: ', accuracy_test)
```

```python
#prediction on test data
y_pred = model.predict(X_test)
```

```python
#comparing predicted vs actual
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head()
```
